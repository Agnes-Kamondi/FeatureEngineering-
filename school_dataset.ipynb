```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(12, 6))
sns.histplot(school_data_df['total_enrollment'], bins=50, kde=True)
plt.title('Distribution of Total Enrollment')
plt.xlabel('Total Enrollment')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=school_data_df, x='school_level', y='student_teacher_ratio')
plt.title('Student-Teacher Ratio by School Level')
plt.xlabel('School Level')
plt.ylabel('Student-Teacher Ratio')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.violinplot(data=school_data_df, x='school_level', y='total_enrollment', inner="quartile")
plt.title('Total Enrollment by School Level')
plt.xlabel('School Level')
plt.ylabel('Total Enrollment')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=school_data_df, x='school_type_text', order=school_data_df['school_type_text'].value_counts().index)
plt.title('Count of Schools by Type')
plt.xlabel('School Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
data = school_data_df.groupby(['survey_year', 'school_level']).size().unstack().fillna(0)

data.plot(kind='bar', stacked=True, figsize=(12, 8))
plt.title('School Level Distribution by Year')
plt.xlabel('Survey Year')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='School Level')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
plt.scatter(
    data=school_data_df,
    x='student_teacher_ratio',
    y='total_females_enrollment',
    s=school_data_df['total_enrollment'] / 100,  
    alpha=0.5
)
plt.title('Student-Teacher Ratio vs. Total Females Enrollment')
plt.xlabel('Student-Teacher Ratio')
plt.ylabel('Total Females Enrollment')
plt.show()
```
