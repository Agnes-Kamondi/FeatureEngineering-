```python
sales_records_df = sales_records_df.drop_duplicates()
```

Removes all duplicate rows from the DataFrame sales_records_df and assigns the cleaned DataFrame back to the same variable. This ensures that only unique rows remain in the DataFrame.


```python
sales_records_df = sales_records_df.dropna()
```

Removes all rows from the DataFrame sales_records_df that contain any missing (NaN) values and assigns the cleaned DataFrame back to the same variable. This ensures that the DataFrame only includes rows with complete data.

```python
sales_records_df['Order Date'] = pd.to_datetime(sales_records_df['Order Date'])
```

Converts the 'Order Date' column in the DataFrame sales_records_df from a string format to a datetime format. This allows for more efficient date-based operations and analysis.

```python
sales_records_df['Ship Date'] = pd.to_datetime(sales_records_df['Ship Date'])
```

Converts the 'Ship Date' column in the DataFrame sales_records_df from a string format to a datetime format. This enables easier manipulation and analysis of shipping dates.

```python
numeric_cols = ['Units Sold', 'Unit Price', 'Unit Cost', 'Total Revenue', 'Total Cost', 'Total Profit']
for col in numeric_cols:
    sales_records_df[col] = pd.to_numeric(sales_records_df[col], errors='coerce')
```

The code converts the columns listed in numeric_cols to numeric data types, setting any non-numeric values to NaN to handle conversion errors. This ensures that all specified columns contain only numeric data for accurate analysis and operations.

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=sales_records_df, x='Item Type', hue='Item Type', palette='viridis', legend=False)
plt.title('Number of Sales for Each Item Type')
plt.xlabel('Item Type')
plt.ylabel('Number of Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

This code creates a bar plot to visualize the distribution of sales across different item types in the sales_records_df DataFrame. It begins by setting the figure size to 10 by 6 inches. The sns.countplot function from Seaborn is used to create the bar plot, with the x-axis representing the 'Item Type' and the hue also set to 'Item Type' to color-code the bars. The 'viridis' color palette is applied, and the legend is disabled for clarity. The plot is then titled "Number of Sales for Each Item Type," with axis labels for 'Item Type' and 'Number of Sales.' The x-axis labels are rotated 45 degrees for better readability, and plt.tight_layout() adjusts the layout to ensure that all elements fit well. Finally, plt.show() displays the plot.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=sales_records_df, x='Total Revenue', y='Total Profit', hue='Item Type', palette='viridis', alpha=0.7)
plt.title('Total Revenue vs. Total Profit')
plt.xlabel('Total Revenue')
plt.ylabel('Total Profit')
plt.legend(title='Item Type')
plt.tight_layout()
plt.show()
```

This code generates a scatter plot to visualize the relationship between total revenue and total profit in the sales_records_df DataFrame. It starts by setting the figure size to 10 by 6 inches. The sns.scatterplot function from Seaborn creates the scatter plot, with 'Total Revenue' on the x-axis and 'Total Profit' on the y-axis. The points are colored according to 'Item Type' using the 'viridis' color palette, and alpha=0.7 makes the points slightly transparent. The plot is titled "Total Revenue vs. Total Profit," with labeled axes for 'Total Revenue' and 'Total Profit.' A legend is added to identify different item types, and plt.tight_layout() ensures that the plot elements fit neatly. Finally, plt.show() displays the scatter plot.

```python
plt.figure(figsize=(12, 8))
sns.violinplot(data=sales_records_df, x='Sales Channel', y='Units Sold', hue='Sales Channel', palette='viridis', legend=False)
plt.title('Distribution of Units Sold by Sales Channel')
plt.xlabel('Sales Channel')
plt.ylabel('Units Sold')
plt.tight_layout()
plt.show()
```

This code creates a violin plot to show the distribution of units sold across different sales channels in the sales_records_df DataFrame. It begins by setting the figure size to 12 by 8 inches. The sns.violinplot function from Seaborn generates the plot, with 'Sales Channel' on the x-axis and 'Units Sold' on the y-axis. The plot uses the 'viridis' color palette to differentiate between sales channels, and the legend is turned off for clarity. The plot is titled "Distribution of Units Sold by Sales Channel," with axis labels for 'Sales Channel' and 'Units Sold.' plt.tight_layout() ensures the plot elements are well-spaced, and plt.show() displays the plot.

```python
corr_matrix = sales_records_df[numeric_cols].corr()

plt.figure(figsize=(12, 8))
sns.heatmap(
    corr_matrix,
    annot=True,            
    cmap='coolwarm',      
    vmin=-1, vmax=1,       
    fmt='.2f',             
    linewidths=0.5,       
    linecolor='black',    
    cbar_kws={'shrink': .8}  
)

plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```

This code generates a heatmap to visualize the correlation between numerical features in the sales_records_df DataFrame. It first calculates the correlation matrix for the columns listed in numeric_cols. Then, it creates a heatmap with a figure size of 12 by 8 inches using sns.heatmap, with annotations for each cell, a 'coolwarm' color map, and correlation values ranging from -1 to 1. The heatmap displays correlations with values formatted to two decimal places, and includes lines between cells (linewidths=0.5) with a black line color. The color bar is slightly shrunk with cbar_kws={'shrink': .8} for better fit. The plot is titled "Correlation Heatmap of Numerical Features," and plt.show() displays the heatmap.

```python
plt.figure(figsize=(12, 8))
sns.boxplot(data=sales_records_df, x='Item Type', y='Total Profit')
plt.title('Distribution of Total Profit by Item Type')
plt.xlabel('Item Type')
plt.ylabel('Total Profit')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

This code creates a box plot to visualize the distribution of total profit across different item types in the sales_records_df DataFrame. It sets the figure size to 12 by 8 inches and uses sns.boxplot to plot 'Item Type' on the x-axis and 'Total Profit' on the y-axis. The plot is titled "Distribution of Total Profit by Item Type," with axis labels for 'Item Type' and 'Total Profit.' The x-axis labels are rotated 45 degrees for better readability, and plt.tight_layout() adjusts the layout to ensure all elements fit well. Finally, plt.show() displays the box plot.
